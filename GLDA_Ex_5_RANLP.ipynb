## SeededLDA 

```python
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
from gensim.parsing.preprocessing import remove_stopwords

import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet

import re
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

import math
from sklearn.feature_extraction.text import TfidfVectorizer


import pyLDAvis
import pyLDAvis.gensim_models
import matplotlib.pyplot as plt
%matplotlib inline
from pprint import pprint
```

```python
cor=data
```

```python
emp_list = []
for i in range(len(cor)):
    for j in range(len(cor[i])):
        if cor[i][j] not in emp_list:
                   emp_list.append(cor[i][j])
emp_list
```

```python
def convert(emp_list):
    return tuple(emp_list)

vocab_format=convert(emp_list)
vocab_format
```

```python
arr_list=[]
for i in cor:
    art_list=[]
    for x in emp_list:
        p=i.count(x)
        art_list.append(p)
    arr_list.append((art_list))
print("arr_list")
```

```python
len(arr_list)
```

```python
a = np.array(arr_list)
```

```python
X = a.astype(np.int32)
```

```python
word2id = dict((v, idx) for idx, v in enumerate(vocab_format))
```

```python
word2id
```

```python
X.shape
```

```python
X.sum()
```

```python
import guidedlda
```

```python
model = guidedlda.GuidedLDA(n_topics=9, n_iter=100, random_state=7, refresh=20)
model.fit(X)
```

```python
topic_word = model.topic_word_
n_top_words = 10
for i, topic_dist in enumerate(topic_word):
    topic_words = np.array(vocab_format)[np.argsort(topic_dist)][:-(n_top_words+1):-1]
    print('Topic {}: {}'.format(i, ' '.join(topic_words)))
```

```python
seed_list=[['electrode', 'poly', 'polymer', 'pedot', 'film'],
['brain', 'network', 'connectivity', 'function', 'region'],
['learning', 'resolution', 'model', 'method', 'proposed'],
['covid19', 'patient', 'disease', 'coronavirus', 'case'],
['balanced', 'idea', 'living', 'organism', 'according']]
seed_arr2=[]
for i in seed_list:
    seed_arr=[]
    for j in i:
        if j in emp_list:
            seed_arr.append(j)
    seed_arr2.append(seed_arr)
seed_arr2
```

```python
# Guided LDA with seed topics.
seed_topic_list =seed_arr2
model = guidedlda.GuidedLDA(n_topics=5, n_iter=100, random_state=7, refresh=20)

seed_topics = {}
for t_id, st in enumerate(seed_topic_list):
    for word in st:
        seed_topics[word2id[word]] = t_id

model.fit(X, seed_topics=seed_topics, seed_confidence=0.15)
```

```python
n_top_words = 10
t_words=[]
topic_word = model.topic_word_
for i, topic_dist in enumerate(topic_word):
    topic_words = np.array(vocab_format)[np.argsort(topic_dist)][:-(n_top_words+1):-1]
    t_words.append(topic_words)
    print('Topic {}: {}'.format(i, ' '.join(topic_words)))
df = pd.DataFrame(t_words)
```

```python
# The document-topic distributions should be retrived as: doc_topic = model.transform(X).
doc_topic = model.transform(X)
for i in range(9):
    print("top topic: {} Document: {}".format(doc_topic[i].argmax(),
                                                  ', '.join(np.array(vocab_format)[list(reversed(X[i,:].argsort()))[0:5]])))
```
